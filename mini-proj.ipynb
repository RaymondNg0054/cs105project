# **Statistical Analysis**

Lets start by examining our first question which is **"Do you believe you have imposter syndrome?"**.

We will first calculate the distribution of our data.

```python
df["83. Do you believe you have imposter syndrome?"].describe()
```

```python
X = df.iloc[:,0].value_counts().drop([0.0])
plt.bar(X.index,X.values, color = 'y')
plt.title("Do you believe you have imposter syndrome?")
plt.xlabel("Self-Assigned Score")
plt.ylabel("Number of Students")
plt.annotate("Neither Agree Nor Disagree",(X.index[2],X.values[2]),textcoords="offset points",xytext=(0,5),ha='center')
plt.annotate("Strongly Agree",(X.index[0],X.values[0]),textcoords="offset points",xytext=(0,3),ha='center')
plt.annotate("Agree",(X.index[1],X.values[1]),textcoords="offset points",xytext=(0,3),ha='center')
plt.annotate("Disagree",(X.index[4],X.values[4]),textcoords="offset points",xytext=(0,5),ha='center')
plt.annotate("Strongly Disagree",(X.index[3],X.values[3]),textcoords="offset points",xytext=(0,5),ha='center')
plt.show()
```

Based on this breakdown, we are able to get a brief summary of participants' self-perceptions. The question, "Do you believe you have imposter syndrome?" was accompanied with a numeric scale from 1-5, where partipants that chose '1' indicated that they strongly disagreed with having imposter syndrome and participants that chose '5' indicated the inverse response of strongly agreement of having imposter syndrome. The average score of these responses was calculated by using the central tendency method of mean which revealed the average response of '3.62'.

Now lets look how why students felt this way about participating.


```python

afraid_count=0
anxious_count=0
okay_count=0
enjoy_count=0
data = df.iloc[:,3]
for string in data:
  if ',' in str(string):
    hold = str(string).split(",")
    for x in hold:
      if x.split()[0]=="Afraid":
        afraid_count+=1
      if x.split()[0]=="Anxious":
        anxious_count+=1
      if x.split()[0]=="Okay":
        okay_count+=1
      if x.split()[0]=="Enjoy":
        enjoy_count+=1
  else:
    if(str(string).split()[0]=="Afraid"):
      afraid_count+=1
    if(str(string).split()[0]=="Anxious"):
      anxious_count+=1
    if(str(string).split()[0]=="Okay"):
      okay_count+=1
    if(str(string).split()[0]=="Enjoy"):
      enjoy_count+=1
sizes = [afraid_count,anxious_count,okay_count,enjoy_count]
colors=['red','green','blue','orange']
labels = ["Afraid of public speaking","Anxious about getting judged","Okay if requirement","Enjoy contributing"]
squarify.plot(sizes =sizes,label=labels,color=colors,alpha=.7)
```

We can see with this graph that the majority of students do not enjoy public speaking when combining the "anxious about getting judged" and "afraid of public speaking" categories.
